```python
import pandas
import matplotlib.pyplot as plt
import seaborn
import numpy as np
import os
import os.path

import matplotlib.ticker as ticker
```

```python
df = pandas.read_csv("RO_results_benchmark_p1.txt", index_col=False, sep = '\t')
```

```python
df['id'] = df['Method'] + '_' + df['LB']
df
```

```python
df_piv = df[df['status'] == 'Optimal'].pivot(index = ['Instance', 'K', 'L', 'VertexBudget', 'ArcBudget', 'Policy'],
        columns = 'id',
        values = 'RO_Objective').reset_index()
df_piv
```

```python
new_is = []
targets = df['id'].unique()
for i, row in df_piv.iterrows():
    append = True
    for j, t1 in enumerate(targets):
        for t2 in targets[j+1:]:
            if ((not pandas.isna(row[t1])) and (not pandas.isna(row[t2]))):
                if row[t1] != row[t2]:
                    append = False
                    
                    print("new", df_piv.loc[i])
    if append:
        new_is.append(i)
```
