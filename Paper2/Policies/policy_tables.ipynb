```python
dats = {}
increcs = [ 5, 10, 30, 'drop']
new_ab = []
df_sorted = df.sort_values('ArcBudget')
for i, row in df_sorted.iterrows():
    index = (row['Instance'],row['K'], row['L'], row['VertexBudget'], row['Policy'])
    if index in dats:
        dats[index] += 1
        #new_ab.append(increcs[dats[index]])
    else:
        dats[index] = 0
    new_ab.append(increcs[dats[index]])
df_sorted['arc_per'] = new_ab
df =  df_sorted
```

```python
df['tottime'] = df['TotalTime2ndS'] + df['TotalTime1stS']
df['nopt'] = df['status'].isin(['Optimal', 'No_matches'])
df['Heuptime'] = round(df['TotalTimeHeu']/df['tottime']*100,2)
df['RecoMIPptime'] = round(df['TotalTimeRecoMIP']/df['tottime']*100,2)
df['RecoCGptime'] = round(df['TotalTimeRecoCG']/df['tottime']*100,2)
df['MP2ndSite'] = (df['TotalIte2ndS'] - df['TotalIteOptP']) - df['TotalIteHeuTrue']
df['MP2ndSptime'] = round(df['TotalTimeMP2ndS']/df['tottime']*100,2)
df['RecoMIPite'] = df['TotalIte2ndS']- df['TotalIteCGTrue']
df['OPTptime'] = df['TotalTimeOptP']/df['tottime']*100
#df['ObjpDet'] = df['RO_Objective']/df['det_obj_val']*100
df['HSPpReco'] = df['2ndS_n_hsp']/df['n_hsp_total']*100
```

```python
df_p = df[df['status'] == 'Optimal']
#Verifying FirstSOnly RO objective value is always at most that of Full
df_p = df_p.pivot(index = ['PDP','K', 'L', 'arc_per', 'ArcBudget','VertexBudget', 'Instance'],
        columns = 'Policy',
        values = 'RO_Objective').reset_index()
#df_p[df_p['FirstSOnly'] >df_p['Full']]
df_p = df_p.dropna()
df_p['delta'] = df_p['Full'] - df_p['FirstSOnly']
df_p[df_p['delta'] < 0]
```

```python
df = df[(df['arc_per'] != 30) & (df['K'] == 4) & (df['PDP'] > 20) & (df['status']  == 'Optimal')]
df = df[df['Policy'] == 'Full']
vals = ['Ite1stS', 'TotalIte2ndS','tottime', 'nopt','MP2ndSptime','Heuptime','RecoMIPptime', 'RecoCGptime', 'MP2ndSite', 'TotalIteHeuTrue', 'RecoMIPite', 'TotalIteCGTrue', 'NumDominatedS', 'OPTptime', 'TotalIteOptP', 'ArcBudget', 'HSPpReco']
df_piv = df.pivot_table(index = ['PDP','K', 'L', 'arc_per','VertexBudget'],
        columns = 'Policy',
        aggfunc = {v:('mean' if v !='nopt' else 'sum') for v in vals},
        values = vals).reset_index()
#df_piv.round(2).to_csv('./test.csv', index = False)
#df_piv.round(2)
```

```python
df_piv.sort_index(axis=1, level=1, inplace=True, sort_remaining = True)
df_piv.columns = [''.join(col) for col in df_piv.columns.values]
#df_piv.round(2).to_csv('./test.csv', index = False)
df_piv.sort_values(['PDP','K','L','ArcBudgetFull'])
df_piv.round(2)
df_piv.columns
df_piv.drop(['K'], axis = 1)
new_cols = {'PDP':'0_PDP', 
            #'K':'1.1_K', 
            'L':'1_L', 
            'VertexBudget':'2_VBudget',
            'ArcBudgetFull':'3_ABudget',
            'noptFull':'4_opt',
            'tottimeFull':'5a_total',
            'HSPpRecoFull':'5_HSP',
            'HeuptimeFull':'6_Heuptime',
            'MP2ndSptimeFull':'7_MP2ndSptime' ,
            'RecoMIPptimeFull':'8_RecoMIPptime', 
            'RecoCGptimeFull':'9_RecoCGptime', 
            'OPTptimeFull':'9a_OPTptime',
            'Ite1stSFull':'9b_Ite1stS', 
            'TotalIte2ndSFull':'9c_Ite2ndS', 
            'TotalIteHeuTrueFull':'9d_HeuTrue', 
            'MP2ndSiteFull':'9e_MP2ndSite', 
            'RecoMIPiteFull':'9f_RecoMIPite',
            'TotalIteCGTrueFull':'9g_CGTrueite',
            'TotalIteOptPFull':'9h_IteOptP', 
            'NumDominatedSFull':'9i_NumDomS',
            #'arc_per':'9j_arc_per'
       }
df_new = df_piv.drop(['K', 'arc_per'], axis = 1).rename(columns=new_cols, inplace=False)
df_new = df_new[sorted(df_new.columns)]
```

```python
idx = pandas.Index(df_new.columns)
table = []
li = []

for j in idx:
    #name = j.split('_')[1]
    li.append(j)
table = []
table.append(li)
df_list = df_new.round(2).to_numpy().tolist()
for obj in df_list:
    new_obj = [int(v) for v in obj[0:4]] + obj[4:]
    table.append(obj)
table
```

```python
latex_table = Texttable()
latex_table.set_cols_align(["c"] * len(table[0]))
latex_table.set_deco(Texttable.HEADER | Texttable.VLINES)
latex_table.add_rows(table)

s = tabulate(table, headers='firstrow', tablefmt='latex')
print(s)

with open('full_table.tex', 'wt') as f:
    f.write(s)
```
